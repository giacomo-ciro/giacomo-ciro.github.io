```python
import json

exam_link_map = {
    "Computer Vision & Image Processing" : "https://didattica.unibocconi.eu/ts/tsn_anteprima.php?cod_ins=20878&anno=2025&IdPag=7921",
    "Deep Learning & Reinforcement Learning": "https://didattica.unibocconi.eu/ts/tsn_anteprima.php?cod_ins=20876&anno=2025&IdPag=7921",
    "Information Theory": "https://didattica.unibocconi.eu/ts/tsn_anteprima.php?cod_ins=20877&anno=2025&IdPag=7921",
}

with open("assets/jsons/timeline.json", "r") as file:
    timeline = json.load(file)

for entry in timeline["timeline"]:
    if entry['event'].startswith('Exam'):
        exam = entry['event'].replace("Exam: ", "").strip()
        if exam in exam_link_map:
            link_to_exam = exam_link_map[exam]
            entry["event"] = f"Exam: <href='{link_to_exam}' target='_blank'>{exam}</a>"

with open("assets/jsons/timeline.json", "w") as file:
    json.dump(timeline, file, indent=4)
```
